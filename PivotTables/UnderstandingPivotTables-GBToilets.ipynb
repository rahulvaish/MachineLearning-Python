### Pivot

```python
girlsPivot = pd.pivot(data=girls, values=['Primary_Only','U_Primary_Only'], index='State_UT', columns='year')
girlsPivot.head(10)
```

```python
print(type(girlsPivot))
```

### Pivot Table

```python
girlsPivotTable = pd.pivot_table(data=girls, values=['Primary_Only','U_Primary_Only'], index='State_UT', columns='year')
girlsPivotTable.head(10)
```

```python
print(type(girlsPivotTable))
```

###### Granular Computing on PIVOT TABLE

```python
df = pd.DataFrame(girlsPivotTable.loc['Andhra Pradesh','Primary_Only'])
df = df.loc['2014-15','Andhra Pradesh']
df
```

```python
df = pd.DataFrame(girlsPivotTable.loc['Andhra Pradesh','Primary_Only'])
df = df['Andhra Pradesh'].mean()
df
```
