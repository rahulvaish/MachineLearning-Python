##  One-Hot Encoding- KDD Way (Nominal Encoding):
<hr>

```python
data=pd.read_csv('C:\\Users\\Lenovo\\Desktop\\IPY\\Complete-Feature-Engineering-master\\mercedesbenz.csv')
data = data[['X1','X2','X3','X4','X5','X6']]
for c in data.columns:
    print("For cloumn",c ," number of unique categories are -> ", len(data[c].unique()))
```

```python
for c in data.columns:
    print("For",c,"the category freqncy is\n", data[c].value_counts().sort_values(ascending=False).head(10))
```

```python
combinedlist=[]
for c in data.columns:
    list = [y for y in data[c].value_counts().sort_values(ascending=False).head(10).index]
    combinedlist.append(list)
```

```python
for i in range(0,len(combinedlist)):   
    print(combinedlist[i])
```

```python
# get whole set of dummy variables, for all the categorical variables

def one_hot_encoding_top_x(data, column, top_categories):
    # function to create the dummy variables for the most frequent labels
    # we can vary the number of most frequent labels that we encode
    
    for label in top_categories:
        data[column+'_'+label] = np.where(data[column]==label, 1, 0)

```

##  One-Hot Encoding- Count/Frequncy Way (Nominal Encoding):
<hr>

```python
data=pd.read_csv('C:\\Users\\Lenovo\\Desktop\\IPY\\Complete-Feature-Engineering-master\\mercedesbenz.csv')
data = data[['X1','X2','X3','X4','X5','X6']]
```

```python
data.head()
```

```python
#for c in data.columns:
    #print("For cloumn",c ," number of unique categories are -> ", len(data[c].unique()))
```

```python
data['X2'].value_counts().sort_values(ascending=False).head()
```

```python
data_frequency_dict = data['X2'].value_counts().to_dict()
```

```python
print(data_frequency_dict)
```

```python
data['X2'] = data['X2'].map(data_frequency_dict)

data.head()
```

##### Count/Frequncy Advantage:
 - It is very simple to implement.
 
##### Count/Frequncy Disadvantage:
 - If some of the labels have the same count, then they will be replaced with the same count and they will loose some valuable information.

##  Label Encoding- System Generated Way (Ordinal Encoding):
<hr>
<p>
    Ordinal scale has all its variables in a specific order, beyond just naming them.
</p>   

```python
data=pd.read_csv('C:\\Users\\Lenovo\\Desktop\\IPY\\Complete-Feature-Engineering-master\\mercedesbenz.csv')
data = data[['X1','X2','X3','X4','X5','X6']]
data.head()
```

```python
# Import label encoder 
from sklearn import preprocessing 
  
# label_encoder object knows how to understand word labels. 
label_encoder = preprocessing.LabelEncoder() 
  
# Encode labels in column 'species'. 
data['X1_ENC1']= label_encoder.fit_transform(data['X1'])
```

```python
data.head()
```

```python
print(data['X1'].unique())
print(len(data['X1'].unique()))
```

```python
print(max(data['X1_ENC1'])) # Encoding start from 0,1,2,3......
```
